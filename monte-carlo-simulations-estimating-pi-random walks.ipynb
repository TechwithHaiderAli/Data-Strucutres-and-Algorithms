```python
import random as r
from math import sqrt
```

### Estimating Pi


```python
def distance(a,b):
    return sqrt(a**2+b**2)

def estiamte_pi(darts):
    inside=0

    for i in range(darts):
        x=r.random()
        y=r.random()

        if distance(x,y)<1.0:
            inside+=1

    pi=(inside/darts)*4

    return pi        
```

```python
num_sim=int(1e+6)
pi=estiamte_pi(num_sim)
print(pi)
```

### Random Walks

```python
import random

def get_action():
    return random.choice([(0,1),(-1,0),(0,-1),(1,0)])
def random_walk(n):
    """Return path after n block random walk"""
    path=[]
    x,y=0,0

    for i in range(n):
        (dx,dy)=get_action() #get step deltas
        x=x+dx
        y=y+dy
        path.append((x,y)) # create a new etuple and save it in the path

    return path    
walk=random_walk(2000)
```

### plotting the Path

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(num=None,figsize=(15,10))

point_color=list(range(2000))
x_numbers=[x[0] for x in walk]
y_numbers=[x[1] for x in walk]

plt.scatter(x_numbers,y_numbers,marker=".",s=100,c=point_color)
```
